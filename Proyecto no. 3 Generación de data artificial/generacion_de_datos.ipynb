```python
#chunk 2
file_path = "datos_credit_scoring.csv"
if os.path.exists(file_path):
    os.remove(file_path)

from faker import Faker
import pandas as pd
import numpy as np

fake = Faker('es_MX')  # Configura Faker para generar datos en español de México

# Configuración inicial
np.random.seed(42)
n_clientes = 100000  # Ajusta este número para el número total de clientes
n_meses = 12

# Generar datos base para cada cliente
ids = np.arange(1, n_clientes + 1)
nombres = [fake.name() for _ in range(n_clientes)]
ocupaciones = [fake.job() for _ in range(n_clientes)]
edades = np.random.randint(18, 70, size=n_clientes)
ingresos = np.random.randint(3000, 20000, size=n_clientes)
creditos_activos = np.random.randint(0, 5, size=n_clientes)
limites = np.random.randint(10000, 50000, size=n_clientes)

# Crear DataFrame base
df = pd.DataFrame({
    'ID': np.repeat(ids, n_meses),
    'NOMBRE': np.tile(nombres, n_meses),
    'AÑO': 2024,
    'MES': np.tile(np.arange(1, n_meses + 1), n_clientes),
    'EDAD': np.tile(edades, n_meses),
    'OCUPACION': np.tile(ocupaciones, n_meses),
    'INGRESO MENSUAL (QTZ)': np.tile(ingresos, n_meses),
    'CANTIDAD DE CREDITOS ACTIVOS': np.tile(creditos_activos, n_meses),
    'LIMITE DE CREDITO (QTZ)': np.tile(limites, n_meses)
})

# Generar variabilidad mensual para "CANTIDAD DE DEUDA (QTZ)" y "PORCENTAJE DEL CRÉDITO (%)"
deudas_base = np.array([limit * np.random.uniform(0.5, 0.9) for limit in limites])
df['CANTIDAD DE DEUDA (QTZ)'] = np.concatenate([np.random.normal(deuda, deuda * 0.05, n_meses) for deuda in deudas_base])
df['PORCENTAJE DEL CRÉDITO (%)'] = df['CANTIDAD DE DEUDA (QTZ)'] / df['LIMITE DE CREDITO (QTZ)'] * 100

# Ajustar los valores y redondear
df['CANTIDAD DE DEUDA (QTZ)'] = df['CANTIDAD DE DEUDA (QTZ)'].round(2)
df['PORCENTAJE DEL CRÉDITO (%)'] = df['PORCENTAJE DEL CRÉDITO (%)'].round(2)

# Guardar en CSV
df.to_csv('data_credit_scoring.csv', index=False)

print("DataFrame generado correctamente y guardado como 'data_credit_scoring_final.csv'.")


print(df.head(24))
```
