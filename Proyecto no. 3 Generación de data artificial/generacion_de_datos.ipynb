```python
# chunk 1
import pandas as pd
from faker import Faker
import numpy as np
import os
import random
```

```python
#chunk 2
file_path = "datos_credit_scoring.csv"
if os.path.exists(file_path):
    os.remove(file_path)


fake = Faker()
nombres = [fake.name() for _ in range(100000)]
ids_personas = {nombre: i for i, nombre in enumerate(nombres, 1)}

ocupaciones = [fake.job() for _ in range(500)]

nombre_ocupacion = {nombre: random.choice(ocupaciones) for nombre in nombres}

ingresos_ocupacion = {ocupacion: random.randint(3000, 20000) for ocupacion in ocupaciones}

meses = list(range(1, 13))  


data = []
for nombre in nombres:
    id_persona = ids_personas[nombre]
    ocupacion = nombre_ocupacion[nombre]
    ingresos = ingresos_ocupacion[ocupacion]
    tarjetas = random.choices([1, 2, 3], weights=[0.5, 0.3, 0.2], k=1)[0]
    limite_credito = ingresos * 3 * tarjetas
    for mes in meses:
        deuda = random.uniform(0.75, 1.15) * ingresos
        pago = random.uniform(0.99, 1.00) * deuda
        data.append({
            'Nombre': nombre,
            'ID Persona': id_persona,
            'Mes': mes,
            'Año': 2025,
            'Ocupación': ocupacion,
            'Ingresos': ingresos,
            'Tarjetas de Crédito': tarjetas,
            'Límite de Crédito': limite_credito,
            'Deuda Total del Mes': round(deuda, 2),
            'Porcentaje Consumido': round(deuda / limite_credito, 2),
            'Pago de la deuda': round(pago, 2)
        })


df = pd.DataFrame(data)

df.sort_values(by=['ID Persona', 'Mes'], inplace=True)

df.to_csv('datos_credit_scoring.csv', index=False)

print(df.head(50))

```
